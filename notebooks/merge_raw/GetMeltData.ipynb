```python
import pandas as pd
import numpy as np
import time
import json
from glob import glob
import requests
from bs4 import BeautifulSoup
import urllib.request
import tarfile
```

```python
# データ取得
def get_data(master_name, coding='utf-8'):
    html_file = f'C:/Users/rin/Downloads/{master_name}.html'
    output_folder = f'../Data/Targz/{master_name}/'
    
    with open(html_file, 'r', encoding=coding) as f:
        html_text = f.read()
    
    soup = BeautifulSoup(html_text, 'html.parser')
    href_list = list(set(href for href in [a.get('href') for a in soup.select('div a')]\
                                          if master_name in href))
    for href in href_list:
        urllib.request.urlretrieve(href, filename=output_folder+href.split('/')[-1])
        # print(f'Downloaded: {href.split("/")[-1]}')
        time.sleep(3)
    print(master_name, 'got!')
```

```python
# データ解凍
def melt_data(master_name):
    needed_file_dict = {
                        'JPWRP': [# 登録マスタ
                                  'upd_mgt_info_p.tsv', 
                                  'upd_right_person_art_p.tsv', 
                                 ], 
                        'JPWIP': [# IPCマスタ
                                 'upd_dsptch_fin_ipc.tsv'
                                 ], 
                        'JPWAP': [# 出願マスタ
                                 'upd_pmac_g_app_case.tsv', 
                                 'upd_sinseinin.tsv'
                                 ]
                        }
    targz_list = glob(f'../Data/Targz/{master_name}/*')
    for targz in targz_list:
        folder_name = targz.split(f'{master_name}_')[-1].split('.')[0]
        with tarfile.open(targz, 'r:gz') as tf:
            member_list = tf.getmembers()
            for member in member_list:
                file_name = member.name.split('/')[-1]
                # print(file_name)
                if file_name in needed_file_dict[master_name]:
                    tf.extract(member, path=f'../Data/Original/{folder_name}/')
                    print(f'Extracted: {folder_name}/{file_name}')
            tf.close()
    print(master_name, 'melted!')
```

```python
for mn in ['JPWRP', 'JPWIP', 'JPWAP']:
    # get_data(mn)
    melt_data(mn)
print('completed!')
```
